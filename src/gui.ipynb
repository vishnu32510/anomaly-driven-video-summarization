```python
import os
import cv2
import tkinter as tk
from tkinter import filedialog, messagebox
from video_summarization import summarize_from_path
from video_summarization_3d import VideoSummarizer

def process_video():
    input_video_path = input_video_entry.get()
    if not input_video_path:
        messagebox.showerror("Error", "Please select a video file.")
        return
    
    summarizer = VideoSummarizer()
    
    # output_video_path = output_video_entry.get()
    # if not output_video_path:
    #     messagebox.showerror("Error", "Please specify an output path for the summarized video.")
    #     return
    print(input_video_path)
    # output_video_path = summarize_from_path(input_video_path)
    output_video_path = summarizer.summarize_from_path(input_video_path)
    print(output_video_path)
    messagebox.showinfo("Success", "Video summarization completed!: ")
    play_video(output_video_path)

def select_input_video():
    filename = filedialog.askopenfilename(filetypes=(("MP4 Files", "*.mp4"), ("All Files", "*.*")))
    input_video_entry.delete(0, tk.END)
    input_video_entry.insert(0, filename)

# def select_output_video():
#     filename = filedialog.askdirectory()
#     output_video_entry.delete(0, tk.END)
#     output_video_entry.insert(0, filename)

def play_video(video_path):
    print(f"Attempting to play video: {video_path}")
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print("Error: Could not open video.")
        messagebox.showerror("Error", "Could not open the video file.")
        return
    
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        cv2.imshow('Summarized Video', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
            
    cap.release()
    cv2.destroyAllWindows()

# Create the GUI
app = tk.Tk()
app.title("Video Summarization Tool")

tk.Label(app, text="Input Video:").grid(row=0, column=0, padx=5, pady=5)
input_video_entry = tk.Entry(app, width=40)
input_video_entry.grid(row=0, column=1, padx=5, pady=5)
tk.Button(app, text="Browse", command=select_input_video).grid(row=0, column=2, padx=5, pady=5)

# tk.Label(app, text="Output Video:").grid(row=1, column=0, padx=5, pady=5)
# output_video_entry = tk.Entry(app, width=40)
# output_video_entry.grid(row=1, column=1, padx=5, pady=5)
# tk.Button(app, text="Save As", command=select_output_video).grid(row=1, column=2, padx=5, pady=5)

tk.Button(app, text="Summarize Video", command=process_video).grid(row=2, column=1, pady=20)

app.mainloop()
```
